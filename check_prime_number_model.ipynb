## The function for checking prime number

- I need to create the function to sort it out.
- The functions the input number whether it is prime number or not.
- By convention, 0 and 1 are not prime.



```python
def checkprime():
    num = int(input('Input the number to check whether it is a prime number: '))
    prime = False
    
    if num == 0 or num == 1:
        prime = False
    
    elif num == 2:
        prime = True
    
    else:
        for i in range(2, num):
            if num % i == 0:
                prime = False
                break
            
        
            while i < num:   
                if num % i != 0:
                    i += 1
                    prime = True
                    continue
                else:
                    prime = False
                    break
    
    if prime:
        print('prime number !!!!!')
    
    else:
        print('It is not a prime number!')
```
