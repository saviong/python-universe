## COUNT PRIMES

- Write a function that returns the number of prime numbers that exist up to and including a given number
- Answer for reference: count_primes(100) --> 25
- When you verify whether it is a prime number, you need to use mod to try all the numbers below the testing one.
- By convention, 0 and 1 are not prime.

```python
def count_primes(x):
    total_prime = 0
    primelist = []
    
    for num in range(2, x+1):
        prime = False

        if num == 0 or num == 1:
            prime = False
        
        elif num == 2:
            prime = True
        
        else:
            for i in range(2, num):
                if num % i == 0:
                    prime = False
                    break
                
            
                while i < num:   
                    if num % i != 0:
                        i += 1
                        prime = True
                        continue
                    else:
                        prime = False
                        break
        
        if prime:
            total_prime += 1
            primelist.append(num)

    return f'The total number of prime numbers that exist up to and including {x} is {total_prime} and the list is {primelist}'
```

```python
count_primes(5)
```

```python
count_primes(100)
```

```python
count_primes(16)
```
