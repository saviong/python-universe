```python
from random import shuffle

game_list = ['','O','','']

def guess_game():
    guess = ''
    shuffle(game_list)
    game = True

    while game == True:
        guess = input("Please pick a number, 0 - 3, to guess the position of 'O': ")
        index = int(guess)
        
        if game_list[index] == 'O':
            print('You guess correct!')
            print(game_list)
            game == False
            break

        else:
            print('Incorrect! Try again please!')
            continue
```
