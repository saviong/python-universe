This is a simple todo list with adding and viewing task function.

```python
print("Hello! Please type 'todo()' to start the todo app and type 'tasks' to get the current todo list.")

tasks = []

def add_task(task):
    tasks.append(task)
    print(f"Added: {task}")

def view_tasks():
    print("\nYour To-Do List:")
    for i, task in enumerate(tasks, 1):
        print(f"{i}. {task}")
    print("\n")

def todo():
    while True:
        print("Options: [1] Add Task  [2] View Tasks  [3] Exit")
        choice = input("Choose an option: ")
    
        if choice == "1":
            task = input("Enter the task: ")
            add_task(task)
            
        elif choice == "2":
            view_tasks()
            
        elif choice == "3":
            print("Goodbye! See you next time!")
            break
            
        else:
            print("Invalid choice, try again.")
```
